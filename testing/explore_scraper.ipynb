```python
from ScraperFC.sofascore import Sofascore
from ScraperFC.transfermarkt import Transfermarkt
import pandas as pd
import re
```

```python
SEASON = "25/26"
LEAGUE = "La Liga"
```

```python
sofa_score = Sofascore()

players_df = sofa_score.scrape_player_league_stats(SEASON, LEAGUE)

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.width', None)

first_10_players = players_df.head(10)

first_10_players.to_excel('data/players_laliga.xlsx', index=False, sheet_name='LaLiga')
first_10_players.to_json('data/players_laliga.json', orient='records', indent=2)

first_10_players
```

```python
with open("data/players_laliga.json", "r") as f:
    data = f.read()

df = pd.read_json(data)

def normalize_column_name(col_name:str) -> str:
    col_name = col_name.replace(" ", "_")
    col_name = re.sub(r'(?<!^)(?=[A-Z])', "_", col_name)
    col_name = col_name.lower()

    return col_name

df = df.rename(columns={col: normalize_column_name(col) for col in df.columns})

df.head(10)
```
